```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy
```

```python
df = pd.read_csv("london_weather.csv") # reading in csv data
print(df.shape)
df # giving out data
```

```python
print("Cloud cover mid:      ", round(numpy.mean(df.cloud_cover),2))
print("sunshine mid:         ", round(numpy.mean(df.sunshine),2))
print("global radiation mid: ", round(numpy.mean(df.global_radiation),2))
print("max temp mid:         ", round(numpy.mean(df.max_temp),2))
print("mean temp mid:        ", round(numpy.mean(df.mean_temp),2))
print("mix temp mid:         ", round(numpy.mean(df.min_temp),2))
print("precipitation mid:    ", round(numpy.mean(df.precipitation),2))
print("Pressure mid:         ", round(numpy.mean(df.pressure),2))
print("snow_depth mid:       ", round(numpy.mean(df.snow_depth),2))
```

```python
sunshine: pd.DataFrame = df.iloc[:,2]
precipatition: pd.DataFrame = df.iloc[:,7]

precipatition.plot(figsize = (30, 18),
              grid=True,
              )
```
